```python
beta = alfa.rename(columns={'Like1': 'price'})
beta
```

```python
beta['price'] = beta['price'].str.replace('$', '')
beta
```

```python
beta['price'] = pd.to_numeric(beta['price'], errors='coerce')
beta['price'] = (beta['price'] * 0.92).round(2)
beta
```

```python
beta['game'] = beta['nounderline'].str.split(':').str[0]
beta['mini'] = beta['nounderline'].str.split(':').str[1]

no_colon = beta['mini'].isna()
beta.loc[no_colon, 'mini'] = beta.loc[no_colon, 'game']
beta.loc[no_colon, 'game'] = None

gamma = beta.drop('nounderline', axis=1)
gamma
```

```python
gamma['faction'] = gamma['game'].apply(
    lambda x: x if str(x).lower() in faction else None
)
gamma
```

```python
mat = pd.read_csv('clear_data/df_materials.csv')
mat
```

```python
# Left join: mantiene todas las filas de mat, agrega precio de gamma
delta = mat.merge(gamma[['mini', 'price']], on='mini', how='left')
delta
```
