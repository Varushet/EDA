```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/DS_Models.csv', sep='|')
df
```

```python
cost = pd.read_csv('data/DS_Model_Costs.csv')
cost
```

```python
df = df.merge(cost[['datasheet_id', 'cost']], on='datasheet_id', how='left')
df
```

```python
cols_to_drop = ['datasheet_id', 'line', 'M', 'T', 'Sv', 'inv_sv', 'inv_sv_descr', 'W', 'Ld', 'OC', 'base_size_descr', 'Unnamed: 13']

alfa = df.drop(columns=[c for c in cols_to_drop if c in df.columns])

alfa['name'].fillna('')
alfa = alfa.dropna(subset=['name'])

alfa = alfa.rename(columns={'name': 'mini'})
alfa
```

```python
alfa = alfa.rename(columns={'base_size': 'base_mm2'})

alfa['base_mm2'] = alfa['base_mm2'].replace(r'^\s*$', np.nan, regex=True)
alfa
```

```python
import re

def clean_mini(name):
    name = str(name)
    name = name.replace('"', "'").replace('"', "'").replace('`', "'").replace('´', "'").replace('’', "'")
    name = re.sub(r'\s+', ' ', name).strip()
    return name

alfa['mini'] = alfa['mini'].apply(clean_mini)

alfa
```

```python
alfa.to_csv('clear_data/df_models.csv', index=False)
```
