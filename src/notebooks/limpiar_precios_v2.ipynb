```python
import pandas as pd
import numpy as np
```

```python
delta = delta.drop(index=delta.index[381:654]).reset_index(drop=True)
delta
```

```python
delta['all'] = delta['faction'].fillna('') + ' ' + delta['mini'].fillna('')
delta['all'] = delta['all'].str.strip()
epsilon = delta.drop(['faction', 'mini'], axis=1)
epsilon
```

```python
def assign_faction(text):
    if pd.isna(text):
        return np.nan
    
    factions = ['Craftworlds',
        'Drukhari',
        'Harlequins',
        'Ynnari',
        'Genestealer_Cult',
        'Tyranids',
        'Necrons',
        'Orks',
        'Tau_Empire',
        'Chaos_Space_Marines',
        'Space_Marines',
        'Supplement_Marines',
        'Blood_Angels',
        'Dark_Angels',
        'Space_Wolves',
        'Deathwatch',
        'Grey_Knights',
        'Astra_Militarum',
        'Adepta_Sororitas',
        'Adeptus_Custodes',
        'Adeptus_Mechanicus',
        'Imperial_Assassins',
        'Imperial_Knights',
        'Inquisition',
        'Chaos_Daemons',
        'Chaos_Knights',
        'Death_Guard',
        'Thousand_Sons']
    
    for f in factions:
        faction_name = f.replace('_', ' ')
        if faction_name in text:
            clean = text.replace(faction_name, '').strip()
            clean = ' '.join(clean.split())
            return f, clean
    return np.nan, text

epsilon[['faction', 'all']] = epsilon['all'].apply(assign_faction).apply(pd.Series)
epsilon
```

```python
faction_map = {
    'Astra Telepathica': 'Adeptus_Custodes',
    'Chaos Space Marine': 'Chaos_Space_Marines',
    'Tallyman': 'Chaos_Daemons',
    'Black Legion': 'Chaos_Space_Marines',
    'Ultramarines': 'Supplement_Marines',
    'Iron Hand': 'Supplement_Marines',
    'Raven Guard': 'Supplement_Marines',
    'Salamanders': 'Supplement_Marines',
    'Imperial Fist': 'Supplement_Marines',
    'White Scar': 'Supplement_Marines',
    'Space Marine': 'Space_Marines',
    'S/Marine': 'Space_Marines',
    'Steel Legion': 'Astra_Militarum',
    'A/I': 'Astra_Militarum',
    'Dark Eldar': 'Drukhari',
    'Eldar Dark': 'Drukhari',
    "T'au": 'Tau_Empire',
    'Gunship': 'Tau_Empire',
    'Tau': 'Tau_Empire',
    'Harlequin': 'Harlequins',
    'Cult': 'Chaos_Daemons',
    'Ork': 'Orks',
    'Genestealers': 'Genestealer_Cult',
    'Tyranid': 'Tyranids',
    'Assassinorum': 'Imperial_Assassins',
    'A/Soror': 'Adepta_Sororitas'
}

name_map = {
    'Daemon': 'Chaos_Daemons',
    'Khorne': 'Chaos_Daemons',
    'Plague': 'Chaos_Daemons',
    'Tzeentch': 'Chaos_Daemons',
    'Kharn': 'Chaos_Daemons',
    'Cultist': 'Chaos_Daemons',
    'Skarbrand': 'Chaos_Daemons',
    'Syll': 'Chaos_Daemons',
    'Feculent': 'Chaos_Daemons',
    'Karanak': 'Chaos_Daemons',
    'Eisenhorn': 'Inquisition',
    'Ravenor': 'Inquisition',
    'Lord Inquisitor': 'Inquisition',
    'Lord Commissar': 'Inquisition',
    'Ciaphas Cain': 'Inquisition',
    'Marneus': 'Supplement_Marines',
    'Primaris': 'Space_Marines',
    'Militarum': 'Astra_Militarum',
    'Cadian': 'Astra_Militarum',
    'Aeronautica': 'Astra_Militarum',
    'Aer': 'Astra_Militarum',
    'Captain General': 'Adeptus_Custodes',
    'Daemons': 'Chaos_Daemons',
    'Chaos': 'Chaos_Daemons',
    'Saint': 'Adepta_Sororitas',
    'Knight': 'Imperial_Knights',
    'Titan': 'Imperial_Knights',
    'Mech': 'Adeptus_Mechanicus',
    'Talons': 'Adeptus_Custodes',
    'Tufts': 'Space_Marines',
    'Space Hulk': 'Space_Marines',
    'Kabalite': 'Drukhari',
    'Commissar': 'Astra_Militarum'
}

import re

def map_faction(text):
    if pd.isna(text):
        return np.nan, text
    
    # faction_map → SÍ elimina la coincidencia de 'all'
    for key, value in faction_map.items():
        pattern = r'\b' + re.escape(key) + r'\b'
        if re.search(pattern, text, re.IGNORECASE):
            clean = re.sub(pattern, '', text, flags=re.IGNORECASE).strip()
            clean = re.sub(r'\s+', ' ', clean).strip()
            return value, clean
    
    # name_map → NO elimina, conserva el texto original
    for key, value in name_map.items():
        pattern = r'\b' + re.escape(key) + r'\b'
        if re.search(pattern, text, re.IGNORECASE):
            return value, text
    
    return np.nan, text

# Aplicar fila por fila solo donde faction es NaN
for idx in epsilon[epsilon['faction'].isna()].index:
    faction, all_text = map_faction(epsilon.loc[idx, 'all'])
    epsilon.loc[idx, 'faction'] = faction
    epsilon.loc[idx, 'all'] = all_text
```
