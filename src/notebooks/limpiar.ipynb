```python
import re

def map_faction(text):
    if pd.isna(text):
        return np.nan, text
    
    # faction_map → SÍ elimina la coincidencia de 'all'
    for key, value in faction_map.items():
        pattern = r'\b' + re.escape(key) + r'\b'
        if re.search(pattern, text, re.IGNORECASE):
            clean = re.sub(pattern, '', text, flags=re.IGNORECASE).strip()
            clean = re.sub(r'\s+', ' ', clean).strip()
            return value, clean
    
    # name_map → NO elimina, conserva el texto original
    for key, value in name_map.items():
        pattern = r'\b' + re.escape(key) + r'\b'
        if re.search(pattern, text, re.IGNORECASE):
            return value, text
    
    return np.nan, text

# Aplicar fila por fila solo donde faction es NaN
for idx in gamma[gamma['faction'].isna()].index:
    faction, _ = map_faction(gamma.loc[idx, 'mini'])
    if pd.isna(faction):
        faction, _ = map_faction(gamma.loc[idx, 'game'])
    gamma.loc[idx, 'faction'] = faction
```
